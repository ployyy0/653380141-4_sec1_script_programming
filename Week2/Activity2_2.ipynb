Activity 2.2: ออกแบบเกมข้อความ "Choose Your Own Adventure"

เกม: การผจญภัยในถ้ำลึกลับ (The Mysterious Cave)

แนวคิดของเรื่อง

ผู้เล่นเป็นนักสำรวจที่พบถ้ำลึกลับ ภายในถ้ำมีหลายเส้นทาง ผู้เล่นต้องเลือกเส้นทางและตัดสินใจว่าจะทำอย่างไร แต่ละการตัดสินใจจะนำไปสู่ผลลัพธ์ที่แตกต่างกัน

จุดตัดสินใจที่ 1

คำถาม
คุณต้องการเลือกเส้นทางใด?

ตัวเลือก
ซ้าย (left)
ขวา (right)
ผลลัพธ์
left → พบหีบสมบัติ
right → พบหมีตัวใหญ่

จุดตัดสินใจที่ 2
กรณีเลือก left

คำถาม
คุณจะเปิดหีบหรือไม่?

ตัวเลือก
yes
no
ผลลัพธ์
yes → ได้รับสมบัติและชนะเกม
no → เดินออกจากถ้ำอย่างปลอดภัย

กรณีเลือก right

คำถาม
คุณจะหนีหรือสู้?
ตัวเลือก
run
fight
ผลลัพธ์
run → หนีออกจากถ้ำได้
fight → หมีชนะ เกมจบ

Pseudo-code
Start

แสดงข้อความต้อนรับ

รับค่า path
(input: left/right)

if path == "left"

    รับค่า choice
    (input: yes/no)

    if choice == "yes"
        แสดง "You found the treasure!"
    elif choice == "no"
        แสดง "You left the cave safely."
    else
        แสดง "Invalid choice."

elif path == "right"

    รับค่า action
    (input: run/fight)

    if action == "run"
        แสดง "You escaped safely."
    elif action == "fight"
        แสดง "The bear defeated you."
    else
        แสดง "Invalid choice."

else

    แสดง "Invalid path."

End

```python
# แสดงข้อความต้อนรับผู้เล่น
print("=== Welcome to the Mysterious Cave ===")

# รับเส้นทางที่ผู้เล่นเลือก
path = input("Choose a path (left/right): ").lower()

# ตรวจสอบว่าผู้เล่นเลือกทางซ้ายหรือไม่
if path == "left":

    # ถ้าเลือกซ้าย ให้ถามว่าจะเปิดหีบหรือไม่
    choice = input("You found a treasure chest! Open it? (yes/no): ").lower()

    # ถ้าตอบ yes
    if choice == "yes":
        print("Congratulations! You found the treasure!")

    # ถ้าตอบ no
    elif choice == "no":
        print("You leave the treasure and walk away safely.")

    # ถ้าพิมพ์อย่างอื่น
    else:
        print("Invalid choice!")

# ถ้าผู้เล่นเลือกทางขวา
elif path == "right":

    # ถามว่าจะหนีหรือสู้
    action = input("A bear appears! Do you run or fight? (run/fight): ").lower()

    # ถ้าเลือกหนี
    if action == "run":
        print("You escaped safely!")

    # ถ้าเลือกสู้
    elif action == "fight":
        print("The bear defeated you. Game Over!")

    # ถ้าพิมพ์อย่างอื่น
    else:
        print("Invalid choice!")

# ถ้าพิมพ์เส้นทางไม่ถูกต้อง
else:
    print("Invalid path!")
```
