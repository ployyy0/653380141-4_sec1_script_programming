```python
temp = 25
is_raining = True

if temp > 30 and is_raining:
    print("Stay inside, it's hot and wet!")
elif temp < 10 or not is_raining:
    print("Go outside and enjoy the weather!")
elif 20 <= temp <= 30 and is_raining:
    print("Bring an umbrella, it's mild but wet.")
else:
    print("Unsure about the weather.")

# Bring an umbrella, it's mild but wet.
```

```python
score = 85

if score >= 80:
    print("Grade: A")
elif score >= 70 :
    print("Grade: B")
elif score >= 60 :
    print("Grade: C")
else:
    print("Grade: C or lower")

# Grade: A
```
