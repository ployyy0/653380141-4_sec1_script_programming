```python
#puzzler3
name = "Ploy"
flag = True

result = name + str(flag)

print(result)
print(type(result))

# เป็น str เพราะ True ถูกแปลงเป็นข้อความด้วย str() ก่อนนำมาต่อกับข้อความ "Ploy" จึงได้ผลลัพธ์เป็นชนิดข้อมูล str
```

```python
#puzzler4
a = True
b = False

result = a and b

print(result)
print(type(result))

# เป็น bool เพราะตัวดำเนินการ and จะให้ผลลัพธ์เป็นค่าความจริง (True หรือ False) เท่านั้น และในกรณีนี้มีค่า False อยู่หนึ่งตัว จึงได้ผลลัพธ์เป็น False (bool)
```

```python
#puzzler5
x = 10
y = 3.5

product = x * y
print(f"Product: {product}")
print(f"Type of product: {type(product)}")

message = "The answer is: " + str(product)
print(message)
print(f"Type of message: {type(message)}")

# product เป็น float เพราะ x เป็น int และ y เป็น float เมื่อคูณกัน (10 * 3.5) Python จะเปลี่ยน int เป็น float ก่อน ทำให้ผลลัพธ์เป็น 35.0 ซึ่งมีชนิดข้อมูลเป็น float

# message เป็น str เพราะใช้ str(product) แปลง 35.0 จาก float เป็นข้อความ แล้วนำไปต่อกับ "The answer is: " จึงได้ผลลัพธ์เป็นข้อความ (str) ทั้งหมด
```
